#### Q2. What city has the highest number of sales?

```python
all_data.head()
```

##### Add a City column

```python
# Use .apply 

all_data['City'] = all_data['Purchase Address'].apply(lambda x: x.split(',')[1]) # lambda x is to work on the cell contents
all_data.head()
```

```python
# We can also write a custom function

def get_city(x):
    return x.split(',')[1]

all_data = all_data.drop(columns='City')
all_data.head()
```

```python
all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x))
all_data.head()
```

```python
# To avoid any duplicates i.e., same city name in 2 states, we will include the state as well

def get_state(x):
    return x.split(',')[2].split(' ')[1]

all_data.drop(columns='City')
all_data['City'] = all_data['Purchase Address'].apply(lambda x: get_city(x) + ' ' + get_state(x))
```

```python
# Use f string

all_data.drop(columns='City')
all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
```

```python
all_data.head()
```

```python
results=all_data.groupby(['City']).sum()
results
```

```python
cities = all_data['City'].unique()

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation='vertical')
plt.ylabel('Sales in USD ($)')
plt.xlabel('City')
plt.show
```

```python
# There is a mismatch between our group by results and the bar chart.
# As per the group by results, San Fransisco has highest Sales.
# As per the bar chart it show Austin with the highest Sales.
# This is due to the fact that the unique values captured and the Sales are not in the same order.


cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, results['Sales'])
plt.xticks(cities, rotation='vertical')
plt.ylabel('Sales in USD ($)')
plt.xlabel('City')
plt.show


# As Data Scientist-->Question-->Why does a particular city has the sales this way? Highest or Lowest or Average. 
# As Business Analyst--> What should be done to improve the Sales in the other cities.
```

### What products are most often sold together?

```python
all_data.head()
```

```python
# Create a new data frame where all sold together orders are kept

df = all_data[all_data['Order ID'].duplicated(keep=False)] #keep is False to all duplicate records
df.head(20)
```

```python
df['Sold Together'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df.head()
```

```python
# Remove duplicates records

df = df[['Order ID','Sold Together']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter
```

```python
count = Counter()

for row in df['Sold Together']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2))) # 2 for most commonly sold 2 items

print(count)
```

```python
# To get more readable format

for key, value in count.most_common(10):
    print(key, value)
```
