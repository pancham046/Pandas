# Filtering

```python
df.loc[df['Type 1'] == 'Grass']
```

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
```

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
new_df # The old index remains in place
```

```python
# new_df = new_df.reset_index() 
# new_df # Adds the old index as a new column

new_df = new_df.reset_index(drop = True) 

# new_df.reset_index(drop = True, inplace = True) # Use the same existing new_df to make changes in it, rather than making changes and then assigning. This will save memory.

new_df
```

```python
# Filter records based on a column containing a particular string

df.loc[df['Name'].str.contains('Mega')]
```

```python
# Filter records based on a column NOT containing a particular string

df.loc[~df['Name'].str.contains('Mega')]
```

```python
# Filter using regex

import re

df.loc[df['Name'].str.contains('pi[a-z]*',flags=re.I,regex=True)] # search all names with letters pi, ingnore the case
df.loc[df['Name'].str.contains('^pi[a-z]*',flags=re.I,regex=True)] # search all names which start with letters pi, ingone the case
```

# Conditional Changes

```python
df.loc[df['Type 1']=='Fire', 'Type 1'] = 'Flamer' # Update the columns where Type 1 is Fire to Flamer
df
```

```python
df.loc[df['Type 1']=='Flamer', 'Type 1'] = 'Fire' # Revert the changes
df
```
